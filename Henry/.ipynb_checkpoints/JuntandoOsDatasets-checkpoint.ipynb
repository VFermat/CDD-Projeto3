# Juntando os datasets

Notebook criado para juntar os datasets criado pelos membros do grupo. O train_dataset.csv e test_dataset.csv contém as features da tranformação de Fourier, Spectral Centroid, Spectral Flatness e Spectral Contrast. O vitinho-train-info.csv e vitinho-test-info contém as features da transformação de Fourier, Harmônicos e Percussão. Os dados foram juntados em um grande dataset nomeado complete_data_train.csv/complete_data_test.csv.

```python
# Módulos necessários para realizar a junção dos datasets:
import pandas as pd
```

```python
# Carregando todos os datasets:
train_dataset = pd.read_csv("train_dataset.csv", header=1)

test_dataset = pd.read_csv("test_dataset.csv", header=1)

vitinho_train = pd.read_csv("vitinho-train-info.csv")

vitinho_test = pd.read_csv("vitinho-test-info.csv")
```

## Juntando os dois datasets com base no train_dataset:

```python
# Criando o dataset base:
complete_data_train = train_dataset

# Colunas a serem movidas:
to_move = ['h1Q', 'h2Q', 'hIQR', 'hmax', 'hmean', 'hmedian', 'hmin', 'hstd', 'htempo', 'p1Q', 'p2Q', 'pIQR', 'pmax', 'pmean',
       'pmedian', 'pmin', 'pstd', 'ptempo']

# Loop para mover todas as colunas:
for column in vitinho_train.columns:
    if column in to_move:
        complete_data_train[column] = vitinho_train[column]

# Mostrando o resultado final:
print(complete_data_train.columns)
```

```python
# Resultado final:
complete_data_train.head(3)
```

## Juntando os dois datasets com base no train_dataset:

```python
# Criando o dataset base:
complete_data_test = test_dataset

# Colunas a serem movidas:
to_move = ['h1Q', 'h2Q', 'hIQR', 'hmax', 'hmean', 'hmedian', 'hmin', 'hstd', 'htempo', 'p1Q', 'p2Q', 'pIQR', 'pmax', 'pmean',
       'pmedian', 'pmin', 'pstd', 'ptempo']

# Loop para mover todas as colunas:
for column in vitinho_test.columns:
    if column in to_move:
        complete_data_test[column] = vitinho_test[column]

# Mostrando o resultado final:
print(complete_data_test.columns)
```

```python
# Resultado final:
complete_data_test.head(3)
```

## Salvando o resultado final

```python
complete_data_train.to_csv(path_or_buf="complete_data_train.csv")

complete_data_test.to_csv(path_or_buf="complete_data_test.csv")
```
